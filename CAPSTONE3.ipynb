```python
from geopy.geocoders import Nominatim
import geocoder
import pandas as pd
```

```python

df=pd.read_csv ("https://raw.githubusercontent.com/isdela/Coursera_Capstone/master/Toronto.csv")
df["Latitude"]=""
df["Longitude"]=""
postal_codes=df["Postcode"]
df.head()
```

```python
for i,row in df.iterrows():
    # initialize your variable to+ None
    lat_lng_coords = None
    j=0
    ok=0
    # loop until you get the coordinates
    while( j<10):
        
        g = geocoder.google('{}, Toronto, Ontario'.format(df.iloc[i,1]))
        lat_lng_coords = g.latlng
        j=j+1
        if lat_lng_coords!=None:
            
            latitude = lat_lng_coords[0]
            df.at[i,"Latitude"]=latitude
            longitude = lat_lng_coords[1]
            df.at[i,"Longitude"]=longitude
            ok=ok+1
            break
print(ok)
```

```python

geolocator = Nominatim(user_agent="ny_explorer")
for i,row in df.iterrows():
    Adress=df.iloc[i,1]
    location = geolocator.geocode(Adress)
    latitude = location.latitude
    longitude= location.longitude
    df.at[i,"Latitude"]=latitude
    df.at[i,"Longitude"]=longitude
df.head()
```

```python
df.to_csv('Toronto_Loc.csv', header=True, index=False)
```
